# Length distribution of key transcript

```python
def to_binned(shape, bins = 100):
    s = np.around(np.linspace(0, len(shape), bins+1))
    new = []
    for i in range(len(s)-1):
        new.append(np.sum(shape[int(s[i]):int(s[i+1])]))
    return np.array(new)
def get_density(key_transcript, read_density, bins = 100):
    all_spect = []
    ids = []
    if len(key_transcript) < 500:
        sample = len(key_transcript)
    else:
        sample = 500
    for i in range(sample):
        chro = key_transcript[i].chrom
        start = int(key_transcript[i].start)
        end = int(key_transcript[i].stop)
        strand = key_transcript[i].strand
    
        shape = read_density.values(chro, start, end, strand)
        if np.nansum(shape)!= 0 and len(shape) > 200:
            #print(chro, start, end, strand)
            if strand == '-':
                shape = [-s for s in shape]
            shape = np.nan_to_num(shape, 0)
            normalized = shape/np.sum(shape)
            all_spect.append(to_binned(normalized, bins = bins))
            ids.append((chro, start, end, strand))
            i+= 1
        else:
            i+= 1
    return all_spect, ids
```

```python
key_transcript_hnrnpc = BedTool('/home/hsher/projects/HNRNPC.key_exon')
all_hnrnpc, id_hnrnpc = get_density(key_transcript_hnrnpc, HNRNPC)
key_transcript_hnrnpk = BedTool('/home/hsher/projects/HNRNPK.key_exon')
all_hnrnpk, id_hnrnpk = get_density(key_transcript_hnrnpk, HNRNPK)
key_transcript_rps3 = BedTool('/home/hsher/projects/RPS3.key_exon')
all_rps3, id_rps3 = get_density(key_transcript_rps3, RPS3)
key_transcript_rps11 = BedTool('/home/hsher/projects/RPS11.key_exon')
all_rps11, id_rps11 = get_density(key_transcript_rps11, RPS11)
```

```python
import seaborn as sns
def plot_individual(all_bins, name = 'HNRNPC'):
    d = np.stack(all_bins)
    sns.heatmap(d, cmap="YlGnBu")
    plt.ylabel('individual intron')
    plt.xlabel('bins')
    plt.title(name)
```
