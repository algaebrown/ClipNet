```python
import matplotlib.gridspec as gridspec

def plot_meta_density(eCLIP, five_utr_len=5000, three_utr_len=10000, intron_len = 150000, exon_len = 10000, ymax = 0.03, logy = False, example = 'positive'):
    ''' line plot of metadensity'''
    fig2 = plt.figure(figsize = (10, 5))
    spec2 = gridspec.GridSpec(ncols=9, nrows=2, figure=fig2)
    
    # axis
    utr_five_1 = fig2.add_subplot(spec2[0, 0])
    exon_1 = fig2.add_subplot(spec2[0, 1:3], sharey = utr_five_1)
    intron_1 = fig2.add_subplot(spec2[0, 3:8], sharey = utr_five_1)
    utr_three_1 = fig2.add_subplot(spec2[0, 8:], sharey = utr_five_1)
    utr_five_2 = fig2.add_subplot(spec2[1, 0], sharey = utr_five_1)
    exon_2 = fig2.add_subplot(spec2[1, 1:3], sharey = utr_five_1)
    intron_2 = fig2.add_subplot(spec2[1, 3:8], sharey = utr_five_1)
    utr_three_2 = fig2.add_subplot(spec2[1, 8:], sharey = utr_five_1)
    
    # trun off yticklabels/ xticklabels
    _ = [plt.setp(ax.get_yticklabels(), visible=False) for ax in [exon_1, exon_2, intron_1, intron_2, utr_three_1, utr_three_2]]
    _ = [plt.setp(ax.get_xticklabels(), visible=False) for ax in [utr_five_1, exon_1, intron_1, utr_three_1]]
    
    # set xlim for exon, intron, utr's maximal length
    _ = [ax.set_xlim(xmin = 0, xmax = three_utr_len) for ax in [utr_three_1, utr_three_2]]
    _ = [ax.set_xlim(xmin = 0, xmax = five_utr_len) for ax in [utr_five_1, utr_five_2]]
    _ = [ax.set_xlim(xmin = 0, xmax = intron_len) for ax in [intron_1, intron_2]]
    _ = [ax.set_xlim(xmin = 0, xmax = exon_len) for ax in [exon_1, exon_2]]
    
    # set ylim for peak height
    _ = [ax.set_ylim(ymin = 0, ymax = ymax) for ax in [utr_five_1, utr_five_2]]
    if logy:
    _ = [ax.set_yscale('log') for ax in [utr_five_1, utr_five_2]]
    
    # plot positive or negative example
    if example == 'positive':
        metagenes = eCLIP.idr_metagene.values()
    else:
        metagenes = eCLIP.neg_metagene.values()
    for m in metagenes:
        # metagene --> uID --> rep1/rep2 --> five_utr/intron/exon/three_utr
        
        
        utr_five_1.plot(m.densities[eCLIP.uID]['rep1']['five_utr'], color = 'grey', alpha = 0.3)
        utr_five_2.plot(m.densities[eCLIP.uID]['rep2']['five_utr'], color = 'grey', alpha = 0.3)
        utr_five_1.set_title('5\' UTR')
    
        exon_1.plot(m.densities[eCLIP.uID]['rep1']['exon'], color = 'grey', alpha = 0.3)
        exon_2.plot(m.densities[eCLIP.uID]['rep2']['exon'], color = 'grey', alpha = 0.3)
        exon_1.set_title('exon')
    
        intron_1.plot(m.densities[eCLIP.uID]['rep1']['intron'], color = 'grey', alpha = 0.3)
        intron_2.plot(m.densities[eCLIP.uID]['rep2']['intron'], color = 'grey', alpha = 0.3)
        intron_1.set_title('intron')
    
        utr_three_1.plot(m.densities[eCLIP.uID]['rep1']['three_utr'], color = 'grey', alpha = 0.3)
        utr_three_2.plot(m.densities[eCLIP.uID]['rep2']['three_utr'], color = 'grey', alpha = 0.3)
        utr_three_1.set_title('3\' UTR')
    plt.suptitle('{} : {} examples'.format(eCLIP.name,example))
```

```python
def bind_stregth_discretize(density_array, bins = 8, ymax = 0.03):
    ''' given density array(sample * length of feature), count discretization'''
    pseudocount = [np.histogram(d[~np.isnan(d)], range = (0, ymax), bins = bins)[0] + 1 for d in density_array.T] # pseudocount 
    den_prob = np.stack([p/np.sum(p) for p in pseudocount]).T
    
    return den_prob

def pos_spec_bind_stregth(metagenes, eCLIP, peak_max = 0.03, bins = 8, align = 'left', five_utr_len=5000, three_utr_len=10000, intron_len = 150000, exon_len = 10000):
    ''' return probability distribution on "binding stregth/normalized peak height" for each position'''
    
    five_utr = []
    intron = []
    exon = []
    three_utr = []
    for transcript_id in metagenes.keys():
        m = metagenes[transcript_id]
        n1, n2 = meta_density(m, eCLIP)
        
        # make to same length (feature_len)
        n1 = [trim_or_pad(n,l, pad_value = np.nan) for (n, l) in zip(n1, [five_utr_len, exon_len, intron_len, three_utr_len])]
        n2 = [trim_or_pad(n,l, pad_value = np.nan) for (n, l) in zip(n2, [five_utr_len, exon_len, intron_len, three_utr_len])]
        
        # save into list
        five_utr.extend([n1[0], n2[0]])
        exon.extend([n1[1], n2[1]])
        intron.extend([n1[2], n2[2]])
        three_utr.extend([n1[3], n2[3]])
    
    # make into np array
    five_utr = np.stack(five_utr)
    three_utr = np.stack(three_utr)
    intron = np.stack(intron)
    exon = np.stack(exon)
    
    # discretize into 10 bins for each position, nan binning
    den_five = bind_stregth_discretize(five_utr, bins = bins, ymax = peak_max)
    den_exon = bind_stregth_discretize(exon, bins = bins, ymax = peak_max)
    den_intron = bind_stregth_discretize(intron, bins = bins, ymax = peak_max)
    den_three = bind_stregth_discretize(three_utr, bins = bins, ymax = peak_max)
    
    return den_five, den_exon, den_intron, den_three

def plot_binding_strength(metagenes, eCLIP, five_utr_len=5000, three_utr_len=10000, intron_len = 150000, exon_len = 10000, bins = 20):
    fig2 = plt.figure(figsize = (20, 5))
    spec2 = gridspec.GridSpec(ncols=10, nrows=1, figure=fig2)
    
    # axis
    utr_five = fig2.add_subplot(spec2[0, 0])
    exon = fig2.add_subplot(spec2[0, 1:3], sharey = utr_five)
    intron = fig2.add_subplot(spec2[0, 3:8], sharey = utr_five)
    utr_three = fig2.add_subplot(spec2[0, 8:9], sharey = utr_five)
    cb = fig2.add_subplot(spec2[0, 9:])
    
    # trun off yticklabels/ xticklabels
    _ = [plt.setp(ax.get_yticklabels(), visible=False) for ax in [exon, intron,utr_three]]
    
    
    # get density
    f,e,i,t = pos_spec_bind_stregth(metagenes, eCLIP, five_utr_len=five_utr_len, three_utr_len=three_utr_len, intron_len = intron_len, exon_len = exon_len, bins = bins)
    
    # get vmax, vmin
    #vmax = np.log(min([np.max(f), np.max(e), np.max(i), np.max(t)]))
    vmax = np.log(1)
    #vmin = np.log(max([np.min(f), np.min(e), np.min(i), np.min(t)]))
    vmin = np.log(0.001)
    
    # greyscale cmap
    cm = sns.cubehelix_palette(50, hue=0.05, rot=0, light=0.9, dark=0, as_cmap = True)
    
    # plot
    sns.heatmap(np.log(i[1:, :]), cmap=cm, ax = intron, cbar = False, vmin = vmin, vmax = vmax) # don't plot bin 0, no information
    sns.heatmap(np.log(e[1:, :]), cmap=cm, ax = exon, cbar = False, vmin = vmin, vmax = vmax)
    sns.heatmap(np.log(f[1:, :]), cmap=cm, ax = utr_five, cbar = False, vmin = vmin, vmax = vmax)
    sns.heatmap(np.log(t[1:, :]), cmap=cm, ax = utr_three, cbar_ax = cb, vmin = vmin, vmax = vmax)
    
    # set label
    cb.set_ylabel('log probability')
    utr_five.set_ylabel('density bins')
    utr_five.set_yticklabels(np.arange(1,bins))
    
    # revert yaxis
    utr_five.invert_yaxis()
    
    # set title
    intron.set_title('intron')
    exon.set_title('exon')
    utr_five.set_title('5\'UTR')
    utr_three.set_title('3\'UTR')
    
    # add name
    plt.suptitle(eCLIP.name)
```

# relative entropy per position


```python
from scipy.stats import entropy
def density_array_entropy(f, fn):
    '''
    relative entropy for each position, f = positive binding density; fn = negative binding examples's density
    '''
    return np.array([entropy(pk = f[1:, pos], qk = fn[1:, pos]) for pos in range(f.shape[1])])

def pos_relative_entropy(metagenes, metagene_neg, eCLIP, five_utr_len=5000, three_utr_len=10000, intron_len = 150000, exon_len = 10000, bins = 20):

    f,e,i,t = pos_spec_bind_stregth(metagenes, eCLIP, five_utr_len=five_utr_len, three_utr_len=three_utr_len, intron_len = intron_len, exon_len = exon_len, bins = bins)
    f_n,e_n,i_n,t_n = pos_spec_bind_stregth(metagene_neg, eCLIP, five_utr_len=five_utr_len, three_utr_len=three_utr_len, intron_len = intron_len, exon_len = exon_len, bins = bins)
    
    f_e = density_array_entropy(f, f_n)
    i_e = density_array_entropy(i, i_n)
    e_e = density_array_entropy(e, e_n)
    t_e = density_array_entropy(t, t_n)
    
    return [f_e, i_e, e_e, t_e]
```

```python
hnrnpc_entropy = pos_relative_entropy(meta_hnrnpc, meta_hnrnpc_neg, HNRNPC,  five_utr_len=500, three_utr_len=500, intron_len = 1500, exon_len = 1000, bins = 20)
rps3_entropy = pos_relative_entropy(meta_rps3, meta_rps3_neg, RPS3, five_utr_len=500, three_utr_len=500, intron_len = 1500, exon_len = 1000, bins = 20)
rbfox2_entropy = pos_relative_entropy(meta_rbfox2, meta_rbfox2_neg, RBFOX2,  five_utr_len=500, three_utr_len=500, intron_len = 1500, exon_len = 1000, bins = 20)
lin28b_entropy = pos_relative_entropy(meta_lin28b, meta_lin28b_neg, LIN28B, five_utr_len=500, three_utr_len=500, intron_len = 1500, exon_len = 1000, bins = 20)
```

```python
plt.figure(figsize=(20,5))
plt.plot(np.concatenate(hnrnpc_entropy), label = 'HNRNPC', color = 'seagreen', alpha = 0.8)
plt.plot(np.concatenate(rps3_entropy), label = 'RPS3', color = 'steelblue',alpha = 0.8)
plt.plot(np.concatenate(lin28b_entropy), label = 'LIN28B', color = 'orchid', alpha = 0.8)
plt.plot(np.concatenate(rbfox2_entropy), label = 'RBFOX2', color = 'orange',alpha = 0.8)
plt.vlines(500, ymax = 1, ymin = 0, color = 'grey')
plt.vlines(2000, ymax = 1, ymin = 0,color = 'grey')
plt.vlines(3000, ymax = 1, ymin = 0, color = 'grey')
plt.legend()
plt.ylabel('relative entropy')
```
