```python
import pandas as pd
df = pd.read_csv('~/projects/allbigwig.tsv', header = None, sep = '\t')
```

```python
bigwigs = df[45]
```

```python
from os import path
import os
outdir = '/home/hsher/seqdata/eclip_raw/'
for b in bigwigs:
    fname = b.split('/')[-1]
    if path.exists(outdir + fname):
        pass
    else:
        print('downloading {}'.format(fname))
        os.system('wget -O {0} {1}'.format(outdir + fname, b))
    
    
```
