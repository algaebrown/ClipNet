## Questions

How many events in RBPs

```python
import os
alt_dir = '/home/hsher/alt_splice/'
folders = [f for f in os.listdir(alt_dir) if '.tar' not in f and '.txt' not in f]
folders.remove('compiled_outputs')
folders.remove('compiled_outputs_copy')

import pandas as pd
events = ['A3SS', 'A5SS', 'MXE', 'RI', 'SE']

eclip_bam = pd.read_csv('/home/hsher/projects/RBP_annot/ENCODE_FINAL_ANNOTATIONS.uidsonly.txt.manifesthg38.txt', sep = '\t', header= 0)

sonar = pd.read_excel('~/projects/RBP_annot/mmc4.xls')
sonar = sonar.drop(0, axis = 0)
def read_folder(folder, event):
    df = pd.read_csv('{}{}/MATS_Norm_output/{}.MATS.JunctionCountOnly.txt'.format(alt_dir,folder,event), sep = '\t', header = 0)
    
    return df

def number_of_events(df):
    fdr005 = df.loc[df['FDR']<0.05]
    fdr01 = df.loc[df['FDR']<0.1]
    fdr02 = df.loc[df['FDR']<0.2]
    
    rbps = [fdr['geneSymbol'].isin(sonar['gene name']).sum() for fdr in [fdr005, fdr01, fdr02, df]]
    
    return [fdr005.shape[0], fdr01.shape[0], fdr02.shape[0], df.shape[0]],rbps
    
    

# in each folder there is MATS_Norm_output/EVENT.MATS.JunctionCountOnly.txt; EVENTS=(A3SS, A5SS, MXE, RI, SE)
```

```python
def run_event(event):
    keys = ['RBP', 'cell line','FDR<0.05','FDR<0.1','FDR<0.2','total', 'RBP_0.05','RBP_0.1', 'RBP_0.2','Total RBP']
    event_df = pd.DataFrame(columns = keys)
    for folder in folders:
        rbp = folder.split('-')[0]
        cell_line = folder.split('-')[2]
        df = read_folder(folder, event)
        no_events, rbps = number_of_events(df)
        event_df= event_df.append(dict(zip(keys, [rbp, cell_line]+no_events + rbps)), ignore_index = True)
    return event_df
```

```python
a3ss_df = run_event('A3SS')
a5ss_df = run_event('A5SS')
mxe_df = run_event('MXE')
ri_df = run_event('RI')
se_df = run_event('SE')
```

```python
import matplotlib.pyplot as plt

def plot_event_dist(df, event):
    max_n_event = df['FDR<0.2'].max()
    df.loc[:, df.columns.str.contains('FDR')].plot.hist(subplots = True, range = (0,max_n_event),bins = 100)
    plt.xlabel('No. events')
    plt.suptitle(event)

    print('RBPs with a lot of event are:')
    print(' '.join(se_df.loc[se_df['FDR<0.2'] > max_n_event/10, 'RBP'].unique().tolist()))


```

```python
def stats(df):
    dstat = [df.loc[:,df.columns.str.contains('FDR')].mean(),
    df.loc[:,df.columns.str.contains('FDR')].max(),
    df.loc[:,df.columns.str.contains('FDR')].median(),
    df.loc[:,df.columns.str.contains('FDR')].min()]
    
    return pd.DataFrame(dstat, index = ['mean', 'max', 'median', 'min'])
    
```

## Do known splicing regulator have more events

```python
rbp_class = pd.read_excel('~/projects/RBP_annot/rbp_class.xlsx', skiprows = 1)
rbp_class.set_index('Unnamed: 0', inplace = True)
rbp_class.drop('Unnamed: 1', axis = 1, inplace = True)
```

```python
splice_regulators = rbp_class.loc[(rbp_class['Splicing regulation'] == 1) | (rbp_class['Spliceosome'] == 1)].index
```

```python
from scipy import stats
def compare_known_unknown(df, event):
    known = df.loc[df['RBP'].isin(splice_regulators)]
    unknown = df.loc[~df['RBP'].isin(splice_regulators)]
    
    max_n_event = df['FDR<0.2'].max()
    f,ax  = plt.subplots(2,1, sharex = True, sharey = True)
    known['FDR<0.2'].plot.hist(ax = ax[0], range = (0,max_n_event), density = True, bins = 100, color = 'grey', alpha = 0.6)
    known['FDR<0.2'].plot.density(ax = ax[0], color = 'red')
    unknown['FDR<0.2'].plot.hist(ax = ax[1], range = (0,max_n_event), density = True, bins = 100, color = 'grey',alpha = 0.6)
    unknown['FDR<0.2'].plot.density(ax = ax[1], color = 'red')
    
    ax[1].set_xlabel('No. events')
    #ax[0].set_ylabel('RBP density')
    
    ax[1].set_title('other RBP')
    ax[0].set_title('known splicing regulators')
    plt.suptitle(event)
    
    
    s = stats.ks_2samp(known['FDR<0.2'], unknown['FDR<0.2'], alternative = 'less')
    print(s)
    ax[1].annotate('pval = {}'.format(s[1]), xy = (100,0))
    
    
```

```python
compare_known_unknown(a5ss_df, 'A5SS')
```

```python
compare_known_unknown(a3ss_df, 'A3SS')
```

```python
compare_known_unknown(mxe_df, 'MXE')
```

```python
compare_known_unknown(se_df, 'SE')
```

```python
compare_known_unknown(ri_df, 'RI')
```

# Percent event in RBP itself
do RBP self regulate?

```python
def event_in_rbp(df):
    subdf = df.loc[df['FDR<0.2']>0]
    portion_02 = subdf['RBP_0.2']/subdf['FDR<0.2']
    portion_total = subdf['Total RBP']/subdf['total']
    
    return portion_02, portion_total
```

```python
def rbp_portion_plot(df, event):
    plt.boxplot(event_in_rbp(df))
    plt.xticks([1,2],['FDR<0.02', 'Total events'])
    plt.ylabel('portion of events target as RBP')
    plt.title(event)
```

```python
rbp_portion_plot(a3ss_df, 'A3SS')
```

```python
rbp_portion_plot(a5ss_df, 'A5SS')
```

```python
rbp_portion_plot(ri_df, 'RI')
```

```python
rbp_portion_plot(se_df, 'SE')
```

```python
rbp_portion_plot(mxe_df, 'MXE')
```
