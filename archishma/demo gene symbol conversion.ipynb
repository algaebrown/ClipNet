# Gene Symbol can be ambiguous. So sometimes you might want /need to convert stuffs into various IDs!

use `mygene` to do this!

```python
import pandas as pd
dep_score = pd.read_csv('D2_DRIVE_gene_dep_scores.csv')
```

```python
dep_score.head()
```

```python
# first we need to split the names out!
gene_symbols = dep_score['Unnamed: 0'].str.split(' ', expand = True)[0]
```

```python
len(gene_symbols)
```

```python
gene_symbols[0] # make sure they are clean, no space, no weird charaters 
```

```python
import mygene
mg = mygene.MyGeneInfo()
converted = mg.querymany(gene_symbols, scopes='symbol',
    fields='ensembl.gene,entrezgene', as_dataframe=True, species = 9606)
```

```python
converted.head() # looks like the number in () are entrezgene id
```
