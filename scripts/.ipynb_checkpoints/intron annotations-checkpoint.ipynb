```python
def generate_id(chunk, first_chunk = False, last_chunk_esnt = None, outfile = '/home/hsher/projects/gencode_cds_annot.gff3'):
    putative_id = chunk[8].str.split(';', expand = True)[0]
    real_id = []
    if not first_chunk:
        esnt = last_chunk_esnt
    else:
        esnt = ''
    for p in putative_id:
        if 'exon' in p:
            real_id.append(p)
            esnt = p[8:]
        else:
            if p[3:] in esnt:
                real_id.append('ID=intron:'+esnt)
            else: 
                real_id.append('ID=intron:'+p[3:]+':0')
    chunk['real_id'] = real_id
    
    if first_chunk:
        with open(outfile, 'w') as f:
            chunk.to_csv(f, '\t', header = False, index = False)
    else:
        with open(outfile, 'a') as f:
            chunk.to_csv(f, '\t', header = False, index = False)
    return esnt
    
```

```python
import pandas as pd
chunks = pd.read_csv('~/projects/gencode_cds_sorted.gff3', sep = '\t', chunksize = 1000, header = None)
i = 0
for c in chunks:
    if i == 0:
        esnt = generate_id(c, first_chunk = True)
    else:
        esnt = generate_id(c, first_chunk = False, last_chunk_esnt = esnt)
    i += 1

```
